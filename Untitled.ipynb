```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_wine
```

```python
winedata=load_wine()
```

```python
wineDataFrame=pd.DataFrame(winedata['data'],columns=winedata['feature_names'])
winedata.target.shape
wineDataFrame['classes']=winedata.target
```

```python
wineDataFrame.head()
```

```python
X=wineDataFrame.iloc[:,:-1]
y=wineDataFrame.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit(X)
```

```python
Scaled_data=scaler.transform(X)
```

```python
Scaled_data
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
pca.fit(Scaled_data)
```

```python
pcatransformed=pca.transform(Scaled_data)
```

```python
Scaled_data.shape
```

```python
pcatransformed.shape
```

```python
Scaled_data
```

```python
pcatransformed
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pcatransformed[:,0],pcatransformed[0:,1],c=winedata['target'])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal component')
```

```python
pca.explained_variance_ratio_
```

```python
pcaDF=pd.DataFrame(pcatransformed,columns=['PCA1','PCA2'])

wineDataFrameTarget=pd.DataFrame(winedata['target'])

final_data=pd.concat([pcaDF,y],axis=1)
final_data
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

linearRegressor=LinearRegression()
mse=cross_val_score(linearRegressor,final_data.iloc[:,:-1],final_data.iloc[:,-1])
mean_mse=np.mean(mse)
mean_mse
```

```python
pc1=final_data.iloc[:,0]
pc2=final_data.iloc[:,1]
sale=final_data.iloc[:,-1]
sale
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge=Ridge()
parameters={'alpha':[1e-15,10,15]}
ridge_regressor=GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(final_data.iloc[:,:-1],final_data.iloc[:,-1])
```

```python
ridge_regressor.best_params_
```

```python
ridge_regressor.best_score_
```
